## 2. Load the data into a DataFrame

For each of the three pages containing data, I make a request and load the results, using pd.json_normalize to unpack the json, then append the data to a DataFrame. Since each loop produces a DF, the index after appending goes 0-9, then 0-9, then 0, so I reset the index for the combined DF to be 0-20.

```python
# for each page with data, make a request & load results, append the data into a dataframe
df = pd.DataFrame()
for i in range(1,4):
    r = requests.get('http://linserv1.cims.nyu.edu:10000/films?_page='+str(i))
    df = df.append(pd.json_normalize(r.json()))

# reset index for combined df
df.reset_index(inplace=True, drop=True)
df
```

## 3. Create Report

I use value_counts to get the counts of movies for each director, and I sort the resulting Series by index (director names, alphabetically) to make it easier to combine my results later for the report.

```python
counts = df.value_counts(['director'])
counts = counts.sort_index()
counts
```

I convert the type of the rt_score column in order to calculate the mean scores, then use groupby to calculate the mean score for each director. I also sort the resulting Series here by index (director name).

```python
# convert type of rt_score column to calculate means
df['rt_score'] = df['rt_score'].astype(int)
df.dtypes # check that type cast worked
```

```python
# calculate mean score for each director, sort by director names
avg_score = df['rt_score'].groupby(df['director'], sort=True).mean()
avg_score.rename('avg_rt_score', inplace=True)
avg_score
```

I can now create the report DataFrame using the results (Series) found above. I convert the avg_rt_score Series into a DataFrame, where the index is already 'director', then add the values from the counts Series to this DataFrame. I then sort the report by avg_rt_score in descending order to match the expected output.

```python
# create the report dataframe using the results
report = pd.DataFrame(avg_score)
report['counts'] = counts.values
# report.index.name = 'director'

# sort report by avg_rt_score in desc order
report = report.sort_values(by=['avg_rt_score'], ascending=False)
report
```
