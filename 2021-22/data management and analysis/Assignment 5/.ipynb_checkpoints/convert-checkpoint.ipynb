## Preprocessing (Cleaning) of Data in top10s.csv

The dataset is pretty clean, but I want to replace zero values in my 'bpm' column as I did previously in Homework 02 and 03. I also want to give the unnamed column a label of "#" to represent the song's ID within this data, which is how it was labeled in the Kaggle page that was the source for the .csv file.

In previous assignments, I only kept some specific columns of interest from the original .csv file (#, title, artist, top genre, bpm), so I will also just keep these columns for this assignment along with the year column for more relevant information about the songs in the dataset.

```python
import pandas as pd
data = pd.read_csv('top10s.csv')
data = data.rename(columns={"Unnamed: 0": "#"})

df = data[['#', 'title', 'artist', 'top genre', 'year', 'bpm']].copy()
df['bpm'] = df.bpm.replace(0, 118)
df
```

The cleaned/transformed data will be stored in a new .csv file called `topsongs2010s.csv`.

```python
df.to_csv('topsongs2010s.csv', index=False)
```
