### 6. describe the kind of buildings that were sold

According to the documentation, in the building class columns, “A” signifies one-family homes, “O” signifies office buildings, and “R” signifies condominiums. I will create a new categorical column in the dataframe that uses the building class codes (in BUILDING SALE, or building class at time of sale) to categorize the properties appropriately (including "other").

```python
import re

def unit_type(code):
    # search for A, O, or R at start of building class code
    if re.match('A.?', code):
        return 'One Family Home'
    elif re.match('O.?', code):
        return 'Office Building'
    elif re.match('R.?', code):
        return 'Condominium'
    else:
       return 'Other'

df['UNIT TYPE'] = df['BUILDING SALE'].apply(unit_type)
df
```

Using this new UNIT TYPE column, I create a visualization comparing the number of each kind of building that was sold. I apply .value_counts on the column to get information how many times each unit type occurs, then plot the result as a bar graph.

```python
df['UNIT TYPE'].value_counts().plot(kind='bar')
```

### 10. which month did the least amount of sales occur


I create a function to apply on the SALE DATE column that uses regular expressions to match the beginning of each date string to the appropriate month. I create a new column in the dataframe that contains the month of the sale.

```python
# using the original dataframe now
import calendar

def month(date):
    if re.match('1/.+', date):
        return 'Jan'
    elif re.match('2/.+', date):
        return 'Feb'
    elif re.match('3/.+', date):
        return 'Mar'
    elif re.match('4/.+', date):
        return 'Apr'
    elif re.match('5/.+', date):
        return 'May'
    elif re.match('6/.+', date):
        return 'Jun'
    elif re.match('7/.+', date):
        return 'Jul'
    elif re.match('8/.+', date):
        return 'Aug'
    elif re.match('9/.+', date):
        return 'Sep'
    elif re.match('10/.+', date):
        return 'Oct'
    elif re.match('11/.+', date):
        return 'Nov'
    elif re.match('12/.+', date):
        return 'Dec'

df['MONTH'] = df['SALE DATE'].apply(month)
df.head()
```

I use .value_counts() again to see which month has the least sales:

```python
df['MONTH'].value_counts()
```

```python
df['MONTH'].value_counts().plot(kind='bar')
```

It appears that August had the least sales by a few hundred sales. This may be partially due it being the end of summer - prices may be high around this time, deterring people from buying homes/properties.
